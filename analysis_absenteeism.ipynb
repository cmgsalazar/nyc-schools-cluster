# Exploring absenteeism rates

Here, we explore whether trends in chronic absenteeism emerge in our clusters.

```python
import pandas as pd
```

```python
df = pd.read_csv("output/clustered-2024.csv")
```

```python
cols_to_use = [ "DBN", "School Name", "Grade", "Year", "% Chronically Absent" ]
attendance_df = pd.read_excel("data/public-school-attendance-results-2019-2024.xlsb", sheet_name="All Students", usecols=cols_to_use)
```

```python
attendance_df = attendance_df[(
    attendance_df["Grade"] == "All Grades") & (
    attendance_df["Year"] == "2023-24"
)]
```

```python
merged_df = df.merge(attendance_df[["DBN", "% Chronically Absent"]], on="DBN", how="left")
merged_df.sample(3)
```

```python
merged_df.to_csv("output/absenteeism.csv", encoding="UTF-8", index=False)
```

```python
trends = merged_df.groupby("kmeans_label")["% Chronically Absent"].agg(["mean", "min", "max", "std"])
trends.T
```
